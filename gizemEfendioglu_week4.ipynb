```python
import bs4
import pandas as pd
import requests
```

```python
response = requests.get('https://sigchi.org/conferences/upcoming-conferences/')
print(response.text) # Print the output
```

```python
soup = bs4.BeautifulSoup(response.text, "html.parser")
print(soup.prettify()) # Print the output using the 'prettify' function
```

```python
# Access the title element
soup.title
```

```python
# Access the title element
soup.title.string
```

```python
# Access data in the first 'p' tag
soup.td
soup.td.string
```

```python
# Retrieve all links in the document (note it returns an array)
soup.find_all('td')
```

```python
# Retrieve elements by class equal to link using the attributes argument
soup.findAll(attrs={'class' : 'mtr-td-tag'})
```

```python
data = soup.findAll(attrs={'class' : 'mtr-td-tag', 'data-mtr-content': 'City'})
print(data[0].string)
print(data[1].string)
print(data[2].string)
print(data[3].string)
```

```python
data = soup.findAll(attrs={'class' : 'mtr-td-tag', 'data-mtr-content': ['Name', 'City', 'Country']})
for i in data:
    print(i.string)
```

```python
f = open('sigchi_data.csv','a') # open new file, make sure path to your data file is correct

p = 0 # initial place in array
l = len(data)-1 # length of array minus one


f.write ('Name, City, Country' + "\n") #write headers

while p < l: # while place is less than length
    f.write(data[p].string + ", " )  # write county and add comma
    p = p + 1 # increment
    f.write(data[p].string + ", ") # write FIPS and add comma
    p = p + 1 # increment
    f.write(data[p].string + ", ") # write Total Pop and add comma
    p = p + 1 # increment
    f.write(data[p].string + "\n") # write Median Income and add comma
    p = p + 1 # increment
  
    
f.close() # close file
```
